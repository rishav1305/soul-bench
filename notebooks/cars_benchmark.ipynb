```python
# Run the full benchmark suite with GPU VRAM tracking
!python scripts/benchmark.py --prompts prompts/ --results-dir results/ --gpu
```

```python
import json
from pathlib import Path

results_dir = Path("results")
for f in sorted(results_dir.glob("*.json")):
    data = json.load(open(f))
    model = data["model"]
    s = data["summary"]
    cars_v = data.get("cars_vram", "N/A")
    print(f"{model}")
    print(f"  Accuracy: {s['avg_accuracy']:.1%}")
    print(f"  Latency:  {s['avg_latency_s']:.1f}s")
    print(f"  CARS_Size: {data['cars_size']}")
    print(f"  CARS_VRAM: {cars_v}")
    if "category_accuracy" in data:
        print(f"  Categories: {data['category_accuracy']}")
    print()
```

## Notes

- **GPU:** T4 (16 GB VRAM) on free-tier Colab
- **Models:** Phi-3.5-mini Q4_K_M, Qwen2.5-3B Q4_K_M
- **Prompts:** 33 total (3 smoke test + 30 across 10 categories)
- **CARS_VRAM** = Accuracy / (VRAM_GB x Latency_s)
